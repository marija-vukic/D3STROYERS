```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import dash
from dash import dcc, html
```

```python
heart_rate = pd.read_csv("HR_001.csv")
heart_rate["datetime"] = pd.to_datetime(heart_rate["datetime"], format="%m/%d/%y %H:%M")
heart_rate["date"] = heart_rate["datetime"].dt.date
heart_rate["hour"] = heart_rate["datetime"].dt.hour
heart_rate["minute"] = heart_rate["datetime"].dt.minute
heart_rate
```

## Comparing Heart Rate and Caloric Intake

```python
heart_rate = heart_rate.groupby(['date','hour','minute']).mean()
heart_rate = heart_rate.reset_index()
heart_rate.head()
```

```python
heart_rate['hr'] = heart_rate[' hr']
```

```python
heart_rate = heart_rate.drop(columns=' hr')
```

```python
heart_rate.head()
```

```python
food = pd.read_csv("Food_Log_001.csv")
food.head()
```

```python
food = food.drop(columns=['time','time_end', 'unit','date'])
```

```python
food["time_begin"] = pd.to_datetime(food["time_begin"] )
food['time'] = pd.to_datetime(food['time_begin']).dt.hour
food['date'] = pd.to_datetime(food['time_begin']).dt.date
food.head()
```

```python
f00d = food[['date', 'time','logged_food','calorie','sugar']]
food = f00d.groupby(['date','time']).mean()
```

```python
food = food.reset_index()
food.head()
```

```python
food.head()
```

```python
# Create the figure
fig = go.Figure()

# Add heart rate as a line plot
fig.add_trace(go.Scatter(
    x=heart_rate["minute"], 
    y=heart_rate["hr"], 
    mode='lines+markers',
    name='Heart Rate (bpm)',
    line=dict(color='blue')
))

# Add calorie intake as a bar chart
fig.add_trace(go.Bar(
    x=food["time"], 
    y=food["calorie"], 
    name='Caloric Intake',
    yaxis='y2',
    opacity=0.6
))

# Update layout to use dual y-axes
fig.update_layout(
    title="Heart Rate and Caloric Intake Over Time",
    xaxis=dict(title="Date and Time"),
    yaxis=dict(title="Heart Rate (bpm)", side="left", showgrid=False),
    yaxis2=dict(title="Caloric Intake", overlaying="y", side="right", showgrid=False),
    legend=dict(x=0, y=1),
    barmode="overlay"
)

# Show interactive plot
fig.show()
```

```python
import plotly.graph_objects as go
import pandas as pd

# Assuming dayonehr contains heart rate data with 'hour', 'minute', and 'hr'
# Convert 'hour' and 'minute' into a continuous time axis (fractional hours)
dayonehr["time"] = dayonehr["hour"] + dayonehr["minute"] / 60

# 1. Downsample heart rate data to 5-minute intervals
dayonehr["time_rounded"] = (dayonehr["minute"] // 5) * 5  # Round minutes to nearest 5
dayonehr_grouped = dayonehr.groupby(["hour", "time_rounded"])["hr"].mean().reset_index()
dayonehr_grouped["time"] = dayonehr_grouped["hour"] + dayonehr_grouped["time_rounded"] / 60

# 2. Apply Rolling Average to smooth heart rate fluctuations
dayonehr_grouped["hr_smoothed"] = dayonehr_grouped["hr"].rolling(window=10, min_periods=1).mean()

# 3. Aggregate calorie intake per hour
# dayonefood_grouped = dayonefood.groupby("hour")["calorie"].sum().reset_index()

# Create the figure
fig = go.Figure()

# 4. Add heart rate as a smooth spline line plot (fixed smoothing to 1.3)
fig.add_trace(go.Scatter(
    x=dayonehr_grouped["time"], 
    y=dayonehr_grouped["hr_smoothed"], 
    mode='lines+markers',
    name='Heart Rate (bpm)',
    line=dict(color='blue', shape='spline', smoothing=1.3),  # Set to max allowed value
    marker=dict(size=2, opacity=0.6)  # Smaller, less prominent markers
))

# 5. Add calorie intake as a bar chart per hour
fig.add_trace(go.Bar(
    x=dayonefood["time"], 
    y=dayonefood["calorie"], 
    name='Caloric Intake',
    yaxis='y2',
    opacity=0.6
))

# Update layout to use dual y-axes and ensure x-axis is in whole hours
fig.update_layout(
    title="Heart Rate and Caloric Intake Over Time",
    xaxis=dict(title="Hour", tickmode="linear", dtick=1),  # Ensure x-axis is in whole hours
    yaxis=dict(title="Heart Rate (bpm)", side="left", showgrid=False),
    yaxis2=dict(title="Caloric Intake", overlaying="y", side="right", showgrid=False),
    legend=dict(x=0, y=1),
    barmode="overlay"
)

# Show interactive plot
fig.show()
```
