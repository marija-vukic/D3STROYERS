```python
import pandas as pd
import numpy as np
```

```python
dexcom_001 = pd.read_csv('Dexcom_001.csv')
dexcom_001 = dexcom_001[12:].reset_index().drop(columns=['index', 'Index'])
dexcom_001 = dexcom_001.rename(columns={'Timestamp (YYYY-MM-DDThh:mm:ss)': 'datetime'})
dexcom_001['datetime'] = pd.to_datetime(dexcom_001['datetime'], format='%Y-%m-%d %H:%M:%S')
dexcom_001['date'] = dexcom_001['datetime'].dt.date
dexcom_001['time'] = dexcom_001['datetime'].dt.time
dexcom_001
```

```python
eda_001 = pd.read_csv('EDA_001.csv')
eda_001['datetime'] = pd.to_datetime(eda_001['datetime'], format='%Y-%m-%d %H:%M:%S')
eda_001['date'] = eda_001['datetime'].dt.date
eda_001['time'] = eda_001['datetime'].dt.time
eda_001 = eda_001.rename(columns={' eda': 'eda'})
eda_001[:100]
```

```python
food_log_001 = pd.read_csv('Food_Log_001.csv')
# food_log_001 = food_log_001.groupby('time_begin').sum().reset_index()
food_log_001 = food_log_001.rename(columns={'time_begin': 'datetime'})
food_log_001['datetime'] = pd.to_datetime(food_log_001['datetime'], format='%Y-%m-%d %H:%M:%S')
food_log_001['date'] = food_log_001['datetime'].dt.date
food_log_001['time'] = food_log_001['datetime'].dt.time
food_log_001['calorie'].dtype
food_log_001
```

```python
hr_001 = pd.read_csv('HR_001.csv')
hr_001['datetime'] = pd.to_datetime(hr_001['datetime'], format='%m/%d/%y %H:%M')

# Reformat to the desired string format
hr_001['datetime'] = hr_001['datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
hr_001['datetime'] = pd.to_datetime(hr_001['datetime'], format='%Y-%m-%d %H:%M:%S')
hr_001 = hr_001.groupby('datetime').mean().reset_index()
hr_001 = hr_001.rename(columns={' hr': 'hr'})
hr_001['date'] = hr_001['datetime'].dt.date
hr_001['time'] = hr_001['datetime'].dt.time
hr_001
```
